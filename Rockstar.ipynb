```python
from tkinter import *
import pandas as pd
import PIL.Image as Image
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from PIL import Image,ImageTk

# Modifications to GUI interface
import tkinter as tk
import os.path 

# Create object 
root = Tk()


dictionary = {}

df = pd.read_csv(r'/path') #replace with your data file path

class Search_Rock:
    Al2O3 = 0
    CaO = 0
    FeO = 0
    K2O = 0
    MgO = 0
    MnO = 0
    Na2O = 0
    SiO2 = 0
    TiO2 = 0
    P2O5 = 0
    C = 0
    def __init__ (self, Al2O3, CaO, FeO, K2O, MgO, MnO, Na2O, SiO2, TiO2, P2O5, C):
        self.Al2O3 = Al2O3
        self.CaO = CaO
        self.FeO = FeO
        self.K2O = K2O
        self.MgO = MgO
        self.MnO = MnO
        self.Na2O = Na2O
        self.SiO2 = SiO2
        self.TiO2 = TiO2
        self.P2O5 = P2O5
        self.C = C

#rename the window's title bar
root.title("RockStar")

# Adjust size 
root.geometry( "800x800" )

# Load an image in the script
img= (Image.open("rockstar.png"))
# Resize the image using resize method
resized_image= img.resize((514,285), Image.ANTIALIAS)
new_image= ImageTk.PhotoImage(resized_image)

# Change the label text 
def show():
    #canvas.delete("all")
    if clicked.get() != "Select Your Rock" :
        rock_properties = dictionary[clicked.get()]
        compound1.set("Al2O3: " + str(rock_properties.Al2O3))
        compound2.set("CaO: " + str(rock_properties.CaO))
        compound3.set("FeO: " + str(rock_properties.FeO))
        compound4.set("K2O: " + str(rock_properties.K2O))
        compound5.set("MgO: " + str(rock_properties.MgO))
        compound6.set("MnO: " + str(rock_properties.MnO))
        compound7.set("Na2O: " + str(rock_properties.Na2O))
        compound8.set("SiO2: " + str(rock_properties.SiO2))
        compound9.set("TiO2: " + str(rock_properties.TiO2))
        compound10.set("P2O5: " + str(rock_properties.P2O5))
        compound11.set("C: " + str(rock_properties.C)) 
        pie_chart()
        
    label.configure( text = clicked.get(), font = ("Arial", 15))
    
    if (os.path.isfile(clicked.get() + ".jpg") == True):
        img2=(Image.open(clicked.get() + ".jpg"))
        resized_rocks= img2.resize((500,500), Image.ANTIALIAS)
        new_rocks= ImageTk.PhotoImage(resized_rocks)
    else:
        img2=(Image.open("default_rock.jpg"))
        resized_rocks= img2.resize((500,500), Image.ANTIALIAS)
        new_rocks= ImageTk.PhotoImage(resized_rocks)
        
    label_image.configure(image=new_rocks)
    label_image.image=new_rocks
        
        
def pie_chart():
    rock_properties = dictionary[clicked.get()]
    pie_Al2O3 = float(rock_properties.Al2O3)*100
    pie_CaO = float(rock_properties.CaO)*100
    pie_FeO = float(rock_properties.FeO)*100
    pie_K2O = float(rock_properties.K2O)*100
    pie_MgO = float(rock_properties.MgO)*100
    pie_MnO = float(rock_properties.MnO)*100
    pie_Na2O = float(rock_properties.Na2O)*100
    pie_SiO2 = float(rock_properties.SiO2)*100
    pie_TiO2 = float(rock_properties.TiO2)*100
    pie_P2O5 = float(rock_properties.P2O5)*100
    pie_C = float(rock_properties.C)*100

    ax= fig.add_subplot(111)

    y = [pie_Al2O3,pie_CaO, pie_FeO, pie_K2O, pie_MgO, pie_MnO, pie_Na2O, pie_SiO2, pie_TiO2, pie_P2O5, pie_C]
    mylabels = ["Al2O3", "CaO", "FeO", "K2O", "MgO","MnO", "Na2O","SiO2", "TiO2","P2O5", "C"]

    ax.pie(y, autopct='%1.1f%%')
    
    canvas = FigureCanvasTkAgg(fig, master = root)
    canvas.draw()
    canvas.get_tk_widget().place(x=350, y=100)
    
fig = Figure(figsize=(5, 5), dpi=100)
    
#column names
namesRocks = df.to_numpy()[:, 0]
namesRocks = namesRocks[:]
namesRocks = namesRocks.tolist()
    
# Dropdown menu options 
options = namesRocks
 
# datatype of menu text 
clicked = StringVar() 
  
# initial menu text 
clicked.set( "Select Your Rock" ) 
  
# Create Dropdown menu 
drop = OptionMenu( root , clicked , *options ) 
drop.place(x=300, y=0)
 
# Create button, it will change label text 
button = Button( root , text = "Enter", font = ("Arial", 10), command = show )
button.place(x=450, y=3)

#pie chart label
TextBox_pie = StringVar()
label_pie = Label(root, text="The composites of the selected rock are").place(x=30, y=50)

# create textbox
compound1 = StringVar()
compound2 = StringVar()
compound3 = StringVar()
compound4 = StringVar()
compound5 = StringVar()
compound6 = StringVar()
compound7 = StringVar()
compound8 = StringVar()
compound9 = StringVar()
compound10 = StringVar()
compound11 = StringVar()

l1 = Label( root , textvariable = compound1, font=("Arial",13), fg = 'blue')
l2 = Label( root , textvariable = compound2, font=("Arial",13), fg = 'orange')
l3 = Label( root , textvariable = compound3, font=("Arial",13), fg = 'green')
l4 = Label( root , textvariable = compound4, font=("Arial",13), fg = 'red')
l5 = Label( root , textvariable = compound5, font=("Arial",13), fg = 'purple')
l6 = Label( root , textvariable = compound6, font=("Arial",13), fg = 'brown')
l7 = Label( root , textvariable = compound7, font=("Arial",13), fg = 'pink')
l8 = Label( root , textvariable = compound8, font=("Arial",13), fg = 'grey')
l9 = Label( root , textvariable = compound9, font=("Arial",13), fg = 'yellow')
l10 = Label( root , textvariable = compound10, font=("Arial",13), fg = 'cyan')
l11 = Label( root , textvariable = compound11, font=("Arial",13), fg = 'blue')
l1.place(x=100, y=100)
l2.place(x=100, y=150)
l3.place(x=100, y=200)
l4.place(x=100, y=250)
l5.place(x=100, y=300)
l6.place(x=100, y=350)
l7.place(x=100, y=400)
l8.place(x=100, y=450)
l9.place(x=100, y=500)
l10.place(x=100, y=550)
l11.place(x=100, y=600)

for index, row in df.iterrows():
    dictionary[row[0]] = Search_Rock(row['Al2O3'],row['CaO'],row['FeO'],row['K2O'],row['MgO'],row['MnO'],row['Na2O'],row['SiO2'],row['TiO2'],row['P2O5'],row['C'])

# Create Label for rock name
label = Label( root , text = " ", font = ("Arial", 15)) 
label.place(x=1000, y=3) 

# Create a label to display the image
label_image = Label(root,image= new_image)
label_image.place(x=950, y=100)

#Execute tkinter 
root.mainloop()
```
